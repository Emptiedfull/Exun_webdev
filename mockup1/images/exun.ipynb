```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read csv file and convert to dataframe

df = pd.read_csv('used_cars_data.csv')
df.dropna()
```

```python
#Print the first twenty records of the dataset.



first_20 = df.head(20) #head returns first n number of rows(20)
print(first_20)
```

```python
#Print the last 20 records of the dataset, in the descending order of ‘year’


Last_20 = df.tail(20)#tail returns last n elements of dataframe
sorted = Last_20.sort_values(by='Year', ascending=False)#sorts in descending order of year

print(sorted)
```

```python
#Display the cars and their characteristics where the ‘Fuel Type’ is CNG.

CNG = df[df['Fuel_Type'] == 'CNG'] #filtering the dataframe where fuel type is CNG
print(CNG)
```

```python
#Count the number of records for the Maruti Wagon R LXI CNG car where ‘Kilometers Driven’ is more than 40000.



MarutiWagon = df[df['Name'] == 'Maruti Wagon R LXI CNG'] #filtering the dataframe where name is Maruti Wagon R LXI CNG
Highusage = MarutiWagon[MarutiWagon['Kilometers_Driven'] > 40000] #filtering the dataframe where kilometers driven is more than 40000

print(Highusage.shape[0]) #shape[0] gives the number of rows in the dataframe


```

```python
#Find out the mean of the ‘Kilometers Driven’ grouped by car model.

mean = (df.groupby('Name'))['Kilometers_Driven'].mean() #grouping by name and calculating mean of kilometers driven

mean
```

```python
#Plot a line chart containing both ‘Mileage’  and ‘Kilometers’ for every car (use different colours and markers for both)
df['Mileage'] = (df['Mileage'].dropna()).astype(int)

df['Kilometers_Driven'] = (df['Kilometers_Driven'].dropna()).astype(int)

plt.plot(df['Kilometers_Driven'],df['Mileage'])

plt.show()
```

```python
#Plot a scatter plot to get the correlation between ‘Engine’ and ‘Power’.

temp = pd.DataFrame()  #creating a temporary dataframe

temp["Engine"] = df['Engine'].str.extract('(\d+)').astype('float')#extracting the numerical part of engine column
temp["Power"] =  df['Power'].dropna().str.extract('(\d+)').astype('float')#extracting the numerical part of power column

 
temp.plot(x = "Power", y = "Engine", kind = "scatter") #plotting scatter plot between engine and power
```

```python
#Make a pie chart for the frequencies of all the cars.

Names = df['Name'].value_counts() #counting the frequency of each car name

plt.pie(Names, labels = Names.index) #plotting pie chart
plt.show()
```

```python
#Plot a horizontal bar graph of ‘seats’ of the last 20 records.

Last_20.plot(x = 'Name', y = 'Seats', kind = 'barh') #plotting horizontal bar graph of seats of last 20 records
```

```python
#Find out which car model has the maximum and which car model has the minimum ‘Price’.

max_price = df['Price'].idxmax() #finding maximum price
min_price = df['Price'].idxmin() #finding minimum price


max_car = df.iloc[max_price] #finding car with maximum price
min_car = df.iloc[min_price]#finding car with minimum price

print(f'maximum priced car is {max_car.Name} at {max_car.Price} rupees')
print(f'minimum priced car is {min_car.Name} at {min_car.Price} rupees')
```
